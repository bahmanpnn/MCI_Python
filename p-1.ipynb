```python
answer=True
while(answer):
    x=int(input("please enter a number= "))
    if(x>=1 and x<=19 and x%2!=0):
        x=(x//2)+1
    
        for i in range(1,x+1):
            print((x-i)*' '+ ((2*i-1)*"*")+2*(x-i)*' '+ ((2*i-1)*"*"))
        
        for i in range(1,x):
            print(i*' '+ ((2*(x-i)-1)*"*")+2*i*' '+ ((2*(x-i)-1)*"*"))

        answer=False
    
```
